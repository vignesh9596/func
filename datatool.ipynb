1: What is NumPy, and why is it widely used in Python?
- A: NumPy is a powerful Python library for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays. It's widely used for scientific computing, data analysis, and machine learning due to its speed and efficiency.

2: How does broadcasting work in NumPy?
- A: Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. Smaller arrays are "broadcast" to match the shape of larger ones without copying data, enabling element-wise operations efficiently.

3: What is a Pandas DataFrame?
- A: A DataFrame is a 2D, labeled data structure in Pandas, similar to a table in SQL or Excel. It consists of rows and columns, each with labels, and supports heterogeneous data types.


4: Explain the use of the groupby() method in Pandas.
- A: groupby() is used to group data based on one or more keys and apply aggregation functions like sum, mean, count, etc. It's essential for segmenting and summarizing data.

5: Why is Seaborn preferred for statistical visualizations?
- A: Seaborn provides a high-level interface for creating attractive and informative statistical plots. It integrates well with Pandas and includes built-in themes and color palettes.


6: What are the differences between NumPy arrays and Python lists?
- A:

NumPy arrays are faster and more memory-efficient.

They support vectorized operations, unlike lists.

Arrays must contain elements of the same data type, while lists can contain mixed types.

7: What is a heatmap, and when should it be used?
- A: A heatmap is a data visualization that uses color to represent values in a matrix. It's useful for showing correlations, missing data, or intensities (e.g., in confusion matrices).


8: What does the term “vectorized operation” mean in NumPy?
- A: Vectorized operations allow operations on entire arrays without explicit loops. They are faster and more concise, leveraging low-level C implementations.


9: How does Matplotlib differ from Plotly?
- A:

Matplotlib is static and more customizable, ideal for publication-quality plots.

Plotly creates interactive, web-based visualizations with zoom, hover, and export features.


10: What is the significance of hierarchical indexing in Pandas?
- A: Hierarchical indexing (MultiIndex) allows multiple levels of indexing on rows or columns. It enables more complex data analysis and reshaping operations like pivoting.

11: What is the role of Seaborn’s pairplot() function?
- A: pairplot() creates a grid of scatter plots for all numerical variable pairs in a dataset, useful for visualizing relationships and distributions.

12: What is the purpose of the describe() function in Pandas?
- A: describe() provides summary statistics (count, mean, std, min, max, etc.) for numeric columns in a DataFrame.

13: Why is handling missing data important in Pandas?
- A: Missing data can bias results or cause errors. Pandas provides tools like dropna(), fillna() to handle such data effectively during preprocessing.


14: What are the benefits of using Plotly for data visualization?
- A:

Interactive charts (zoom, hover, export)

Web compatibility (HTML, dashboards)

Easy to use with Pandas and supports 3D and map visualizations


15: How does NumPy handle multidimensional arrays?
- A: NumPy supports n-dimensional arrays (ndarray) and provides tools to reshape, slice, and perform operations across multiple axes efficiently.

Bokeh in Data Visualization

16: What is the role of Bokeh in data visualization?
- A: Bokeh is a Python library for creating interactive and real-time visualizations for web browsers, ideal for dashboards and streaming data.

17: Explain the difference between apply() and map() in Pandas.
- A:

map() is used for element-wise transformations on a Series.

apply() can be used on both Series and DataFrames for applying functions to rows/columns.

18: What are some advanced features of NumPy?
- A:

Broadcasting

Vectorized operations

Masking and filtering

Linear algebra (linalg)

Fourier transforms and random number generation

19: How does Pandas simplify time series analysis?
- A: Pandas supports date/time indexing, resampling, rolling window operations, and time zone handling, making time series analysis more accessible and efficient.

20: What is the role of a pivot table in Pandas?
- A: Pivot tables summarize data by grouping and aggregating values across multiple axes, useful for multidimensional data analysis.

21: Why is NumPy’s array slicing faster than Python’s list slicing?
- A: NumPy slicing is faster because it returns views (not copies) and uses optimized C-based memory access, whereas Python lists require element-by-element copying.

22: What are some common use cases for Seaborn?
- A:

Visualizing distributions (distplot, kdeplot)

Categorical plots (boxplot, violinplot)

Correlation heatmaps

Regression analysis (lmplot)

Pairwise relationships (pairplot)

```python
#1 How do you create a 2D NumPy array and calculate the sum of each row)
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print("Row sums:", row_sums)
```

```python
#2 Write a Pandas script to find the mean of a specific column in a DataFrameA
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [5, 15, 25]})
mean_B = df['B'].mean()
print("Mean of column B:", mean_B)
```

```python
#3  Create a scatter plot using MatplotlibA
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [5, 6, 7, 8]
})

corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
 # 5 Generate a bar plot using PlotlyA
 import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Orange'], 'Quantity': [10, 15, 7]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Fruits', y='Quantity', title='Fruit Quantities')
fig.show()
```

```python
# 6  Create a DataFrame and add a new column based on an existing columnA
df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Score': [80, 90]})
df['Grade'] = df['Score'].apply(lambda x: 'Pass' if x >= 85 else 'Fail')
print(df)
```

```python
# 7  Write a program to perform element-wise multiplication of two NumPy arraysA
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise multiplication:", result)
```

```python
# 8  Create a line plot with multiple lines using MatplotlibA
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 18, 22, 28]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title('Multiple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# 9 Generate a Pandas DataFrame and filter rows where a column value is greater than a thresholdA
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 20]})
filtered_df = df[df['Age'] > 22]
print(filtered_df)
```

```python
#10 Create a histogram using Seaborn to visualize a distributionA
import seaborn as sns

data = [10, 20, 20, 30, 30, 30, 40, 40, 50]
sns.histplot(data, kde=True, bins=5)

```

```python
 # 11 Perform matrix multiplication using NumPyA
 a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
product = np.dot(a, b)
print("Matrix multiplication:\n", product)
```

```python
# 13  Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 20, 25, 30],
    'z': [5, 15, 10, 20]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
